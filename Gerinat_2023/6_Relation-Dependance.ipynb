```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import locale
from datetime import datetime
import warnings
warnings.simplefilter("ignore")#, UserWarning
import datetime
import seaborn as sns
```

```python
Names=['Times','Pression1']
Type ={'Pression': float}#,dtype=Type
dfTS=pd.read_csv('Data/Simulate/SimulatedTrendSaisonality.csv',dtype=Type,header=0,skiprows=0, parse_dates=['Times'],names=Names,index_col=0)
```

```python
Names=['Times','Pression2']
Type ={'Pression': float}#,dtype=Type
df2TS=pd.read_csv('Data/Simulate/SimulatedTrendSaisonalityRelation.csv',dtype=Type,header=0,skiprows=0, parse_dates=['Times'],names=Names,index_col=0)
```

```python
fig, ax = plt.subplots()
ax.scatter(dfTS.values,df2TS.values)

ax.set(xlabel=' série 1', ylabel=' série 2',
       title='Nuage de points correspondant à la série 1 contre al série 2 ')
ax.grid()

fig.savefig("plot_relation_nuage.png")
plt.show()
```

```python
 import statsmodels.api as sm
```

```python
Y = df2TS['Pression2']

X = dfTS['Pression1']
X = np.vander(X, 2) #ici on ajoute
#X = sm.add_constant(X)
model = sm.OLS(Y,X)
results = model.fit()
results.params
```

```python
print(results.summary())
```

```python
results.t_test([1])
```
