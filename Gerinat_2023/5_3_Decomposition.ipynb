```python
import statsmodels.api as sm
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Names=['Times','Pression']
Type ={'Pression': float}#,dtype=Type
dfSimu=pd.read_csv('Data/Simulate/SimulatedTrendSaisonality.csv',dtype=Type,header=0,skiprows=0, parse_dates=['Times'],names=Names,index_col=0)
dfSimu.plot()
```

c'est la synthése du trvail fait précédement ou il existe des méthodes pour décomposer automatiquement une série temporelle.

La bibliothèque statsmodels fournit une implémentation de la méthode de décomposition naïve ou classique dans une fonction appelée season_decompose() . Il nécessite que vous spécifiiez si le modèle est additif ou multiplicatif.

Les deux produiront un résultat et vous devez faire attention à être critique lors de l'interprétation du résultat. Un examen d'un graphique de la série chronologique et de quelques statistiques récapitulatives peut souvent être un bon début pour savoir si votre problème de série chronologique semble additif ou multiplicatif.



```python
from statsmodels.tsa.seasonal import seasonal_decompose
res = seasonal_decompose(dfSimu["Pression"].values.ravel(), period=96,two_sided=False)
plt.plot(res.seasonal,label='seasonnalité')
plt.plot(res.trend,label="trend")
plt.legend()
plt.show()
```
